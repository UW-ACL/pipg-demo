\begin{align*}
\texttt{rd2o} &= \frac{\|z-z^\star\|_2^2}{\|z^\star\|^2_2},\\
\texttt{rfval} &= \frac{|f(z^\star)-f(z)|}{|f(z^\star)|},\\
\texttt{rd2K} &= \frac{\|Hz-g-\prod_{\mathbb{K}}\left[Hz-g\right]\|_2^2}{\|z^\star\|^2_2}.
\end{align*}
