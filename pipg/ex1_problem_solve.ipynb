Settling point of rd2o is determined by accuracy of the JuMP solution.

```

\begin{align*}
\texttt{rd2o} &= \frac{\|z-z^\star\|_2^2}{\|z^\star\|^2_2},\\
\texttt{rfval} &= \frac{|f(z^\star)-f(z)|}{|f(z^\star)|},\\
\texttt{rd2K} &= \frac{\|Hz-g-\prod_{\mathbb{K}}\left[Hz-g\right]\|_2^2}{\|z^\star\|^2_2}.
\end{align*}

```

        Q = diag( 0.5*ones(nx/2) , 2*ones(nx/2) )
        R = diag( 0.1*ones(nu) )
It performs *poorly* with 
        Q = diag( 0.5*ones(nx/2) , 2*ones(nx/2) )
        R = diag( 5.0*ones(nu) )
